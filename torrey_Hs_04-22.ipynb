```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the new CSV file
file_path = 'torrey_outer_Hs.csv'
data = pd.read_csv(file_path, index_col=0, parse_dates=True)

# Resample the data by month and find the maximum wave height for each month
monthly_max_wave_height = data['waveHs[unit="meter"]'].resample('M').max()

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(monthly_max_wave_height, marker='o', linestyle='-', color='b')
plt.title('Monthly Maximum Wave Height (2001-2023)')
plt.xlabel('Date')
plt.ylabel('Wave Height (meters)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'torrey_outer_Hs.csv'
data = pd.read_csv(file_path, index_col=0, parse_dates=True)

# Resample the data by month and find the maximum wave height for each month
monthly_max_wave_height = data['waveHs[unit="meter"]'].resample('M').max()

# Calculate the wave energy by squaring the wave height
monthly_wave_energy = monthly_max_wave_height ** 2

# Calculate the mean wave energy
mean_wave_energy = monthly_wave_energy.mean()

# Calculate the Mean Absolute Deviation (MAD)
mean_absolute_deviation = (monthly_wave_energy - mean_wave_energy).abs()

# Plot the MAD graph with the mean line
plt.figure(figsize=(12, 6))
plt.plot(monthly_wave_energy.index, mean_absolute_deviation, 
         marker='o', linestyle='-', color='r', label='MAD')
plt.axhline(mean_wave_energy, color='g', linestyle='--', label='Mean Wave Energy')
plt.title('Mean Absolute Deviation (MAD) and Mean of Monthly Maximum Wave Energy Torrey Offshore (2001-2023)')
plt.xlabel('Date')
plt.ylabel('Value(square meters)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'torrey_outer_Hs.csv'
data = pd.read_csv(file_path, index_col=0, parse_dates=True)

# Resample the data by month and find the maximum wave height for each month
monthly_max_wave_height = data['waveHs[unit="meter"]'].resample('M').max()

# Calculate the wave energy by squaring the wave height
monthly_wave_energy = monthly_max_wave_height ** 2

# Calculate the mean wave energy
mean_wave_energy = monthly_wave_energy.mean()

# Calculate the deviation from the mean
deviation_from_mean = monthly_wave_energy - mean_wave_energy

# Plot the deviations from the mean
plt.figure(figsize=(12, 6))
plt.plot(monthly_wave_energy.index, deviation_from_mean, 
         marker='o', linestyle='-', color='purple', label='Deviation from Mean')
plt.axhline(0, color='g', linestyle='--', label='Mean')
plt.title('Deviation from Mean of Monthly Maximum Wave Energy Torrey Offshore (2001-2023)')
plt.xlabel('Date')
plt.ylabel('Deviation from Mean (square meters)')
plt.legend()
plt.grid(True)
plt.show()
```
